```python
import spacy
import pymorphy2

with open("GenП_Chapter_2_rus.txt", 'r', encoding="UTF-8") as f:
    text = f.read()

text = text.replace('\n', ' ')

nlp = spacy.load("ru_core_news_lg")
doc = nlp(text)
morph = pymorphy2.MorphAnalyzer()


def pymorphy_lemmatizer(token):
    parses = morph.parse(token.text)
    if parses:
        normal_form = parses[0].normal_form
        return normal_form
    else:
        return token.lemma_
```

```python
pers = ["татарский", "сергей", "морковин", "морковина", "вова", "гусейн"]
ents = doc.ents
names = list()
for ent in ents:
    if ent.label_ == "PER" and  pymorphy_lemmatizer(ent) not in pers and len(str(ent)) > 3:
            names.append(pymorphy_lemmatizer(ent))
print(names)
```

```python
from spacy import displacy

html = displacy.render(doc, style="ent")

with open ("pelebin_hz_chap2.html", "w", encoding="UTF-8") as f:
    f.write(html)
```
