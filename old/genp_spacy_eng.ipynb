```python
import spacy
import re

with open("GenП_Chapter_2_eng.txt", 'r') as f:
    text = f.read()

def is_valid_entity(entity):
    if re.search(r'[^\w\s]', entity): 
        return False
    if any(char.isdigit() for char in entity):
        return False
    if len(entity) <= 1:
        return False
    return True

nlp = spacy.load("en_core_web_md")
cleaned_text = text.replace('"', '"').replace("“", '"').replace("”", '"')
doc = nlp(cleaned_text)
```

```python
pers = ["Tatarsky", "Morkovin", "Smirnoff", "Vova", "Hussein", "Sergei", "Chechen", "Lena", "Azadovsky"]
sentences = list(doc.sents)
ents = list(),
names = list()
for ent in doc:
    if ent.ent_type_ == "PERSON" and str(ent) not in pers:
            names.append(ent)
print(names)
```
